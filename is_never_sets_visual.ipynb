```python
import datetime

import pygmt

fig = pygmt.Figure()
# Create a figure showing the global region on a Mollweide projection
# Land color is set to dark green and water color is set to light blue
fig.coast(region="d", projection="W0/15c", land="darkgreen", water="lightblue")
# Set a time for the day-night terminator and twilights
# terminator_datetime = datetime.datetime(
#     year=2000, month=1, day=1, hour=17, minute=0, second=0
# )
terminator_datetime = datetime.datetime.now(tz=datetime.timezone.utc)
# Set the pen line to be 0.5 points thick
# Set the fill for the night area to be navy blue at different transparency
# levels
fig.solar(
    terminator="day_night",
    terminator_datetime=terminator_datetime,
    fill="navyblue@20",
    pen="0.5p",
)
fig.solar(
    terminator="civil",
    terminator_datetime=terminator_datetime,
    fill="navyblue@15",
    pen="0.5p",
)
fig.solar(
    terminator="nautical",
    terminator_datetime=terminator_datetime,
    fill="navyblue@10",
    pen="0.5p",
)
fig.solar(
    terminator="astronomical",
    terminator_datetime=terminator_datetime,
    fill="blue@5",
    pen="0.5p",
)


##add markers at Paris and show label(with offset)
fig.plot(x=[2.35], y=[48.85], style="c0.2c", fill="red", pen="black")

fig.text(
    x=2.35,
    y=48.85,
    text="Paris",
    font="12p,Helvetica-Bold,black",
    justify="LM",
    offset="0.2c",
)

##add markers at Beijing and show label(with offset)
fig.plot(x=[116.4], y=[39.9], style="c0.2c", fill="red", pen="black")

fig.text(
    x=116.4,
    y=39.9,
    text="Beijing",
    font="12p,Helvetica-Bold,black",
    justify="LM",
    offset="0.2c",
)

##add markers at Sanfrancisco and show label(with offset)
fig.plot(x=[-122.4], y=[37.8], style="c0.2c", fill="red", pen="black")

fig.text(
    x=-122.4,
    y=37.8,
    text="Sanfrancisco",
    font="12p,Helvetica-Bold,black",
    justify="LM",
    offset="0.2c",
)

fig.show()
```
